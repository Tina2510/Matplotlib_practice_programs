3rd que [Write a Python program to draw a line using given axis values taken from a text file, with 
suitable label in the x axis, y axis and a title. ]

```python
x = []
y = []
with open("test.txt" ,"r") as text:
    for data in text:
        a, b = data.split()
        x.append(int(a))
        y.append(int(b))

plt.plot(x, y)
plt.xlabel("x points")
plt.ylabel("y points")
plt.title("graph")
plt.show()       
```

4th[draw line charts of the financial data of Alphabet Inc]

```python
import pandas as pd
import matplotlib.pyplot as plt

data = """Date,Open,High,Low,Close
03-10-16,774.25,776.065002,769.5,772.559998
04-10-16,776.030029,778.710022,772.890015,776.429993
05-10-16,779.309998,782.070007,775.650024,776.469971
06-10-16,779,780.47998,775.539978,776.859985
07-10-16,779.659973,779.659973,770.75,775.080017
"""

with open("fdata.csv", "w") as file:
    file.write(data)
df = pd.read_csv("fdata.csv")

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')

plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Alphabet Inc")
plt.legend()
plt.show()
```

12 Write a Python program to create multiple types of charts

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

names = ['A', 'B', 'C', 'D']
score = [25, 10, 35, 25]
plt.plot(x, y)
plt.title("Line Chart")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
plt.bar(names, score)
plt.title("Bar Chart")
plt.xlabel("Names")
plt.ylabel("Score")
plt.show()
plt.scatter(x, y)
plt.title("Scatter Chart")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
plt.pie(score, labels=names)
plt.title("Pie Chart")
plt.show()
```

14[ Customized the grid lines with 
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks]

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("fdata.csv")
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')
plt.plot(df['Date'], df['Close'])
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Alphabet Inc")
plt.minorticks_on()
plt.grid(True, which='major', linewidth=0.8)
plt.grid(True, which='minor', linewidth=0.4)
plt.tick_params(axis='both', which='both', length=0)

plt.show()
```
